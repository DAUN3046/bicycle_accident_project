```python
# 대구의 중상~사망 사고 비율 나타낸 히트맵

import pandas as pd
import folium
from folium.plugins import HeatMap

df = pd.read_excel('2012.xlsx')

# define centre of map
TOR_COORDINATES = (df['위도'].mean(), df['경도'].mean())
 
# subset to match subset of locations
MAX_RECORDS = 60000
  
# create empty map zoomed in on Toronto
map_tor = folium.Map(location=TOR_COORDINATES, zoom_start=12)
display(map_tor)
```

```python
# 대구의 중상~사망 사고 비율 나타낸 히트맵

from folium.plugins import HeatMap

# define centre of map
TOR_COORDINATES = (df['위도'].mean(), df['경도'].mean())
 
# subset to match subset of locations
MAX_RECORDS = 60000
  
# create empty map zoomed in on Daegu
map_tor = folium.Map(location=TOR_COORDINATES, zoom_start=12)

# define heat map

HeatMap(data=df[['위도', '경도', 'dangerrate']].groupby(['위도', '경도']).sum().reset_index().values.tolist(),
        radius=8, max_zoom=13).add_to(map_tor)


display(map_tor)

# 출처 : https://rfriend.tistory.com/419
```

## 교차로에서 사고가 많이 남.

```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')
plt.title('자전거 교통사고 특성')
gahaeja=[-13.1,-16.3,-16.7,-23.3]
pihaeja=[11.7,16.0,15.7,24.2]
plt.barh(range(4),gahaeja,label='가해자')
plt.barh(range(4),pihaeja,label='피해자')
plt.legend()
plt.show()
```

# 전국 folium heatmap

```python
import folium
import pandas as pd
from folium.plugins import HeatMap

df = pd.read_excel('1219biaccident.xlsx')

# define centre of map
TOR_COORDINATES = (df['위도'].mean(), df['경도'].mean())
 
# subset to match subset of locations
MAX_RECORDS = 60000
  
# create empty map zoomed in on Toronto
map_entire = folium.Map(location=TOR_COORDINATES, zoom_start=8)

# define heat map

HeatMap(data=df[['위도', '경도']].groupby(['위도', '경도']).sum().reset_index().values.tolist(), radius=8, max_zoom=13).add_to(map_entire)


display(map_entire)
```
